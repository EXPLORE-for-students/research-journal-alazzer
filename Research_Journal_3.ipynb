# 3: So what? (What does it mean?)
## Describe your results

The calculated final radial velocities of HV7’s frames showed that frames 1 and 4 were fairly consistent, but frame 3 produced a significantly lower RV than frames 1 & 4. This could be due to the full-width half-maximum measurements of the absorption features, since frame 3’s FWHM was smaller than the other two frames. Another reason for the difference could be that the signal-to-noise, thus the overall quality, of frame 3’s spectrum was less than frames 1 & 4.

Compared to HV9, the quality of HV7’s spectra were noticeably improved, as the Hα absorption lines were clearer, and the final shifted and stacked spectrum was far closer to the rest wavelength of Hα than HV9’s. One main reason for this is the poorer quality of the data, as the average signal-to-noise ratio of the frames for HV7 was more than double that of HV9.

The low quality of the data, particularly of HV9, can be seen by looking at the inconsistency across frames in the table below, which compares the FWHM and final radial velocity values. From this table, it is noticeable that the frames are more consistent for HV7 than HV9. The FWHM of HV9’s frames vary by 0.9, whereas HV7’s frames vary by 0.13. A large variation in FWHM can indicate weak/poorly defined spectral lines. In a good-quality data set, we would expect the absorption features to look similar and have measurements comparable across multiple frames. The large difference in FWHM makes taking accurate wavelength measurements difficult. Although the overall range of velocities is larger for HV7, the best spectra (frames 1 and 4) are very consistent with each other (difference less than 4 km/s). In comparison, the two closest RV values for HV9 were almost 10 km/s apart, which is a greater difference than aimed for.

A hypothesis our group had for why the spectral quality of HV9 was poor was that the observation would have been taken near the minimum light of the star. Since RR Lyrae are variable stars, they pulsate, meaning they vary in brightness over a certain period, which can be observed. Minimum light refers to the point in the star's period where the brightness is at its lowest. Since the phases were calculated using a maximum light time as the adopted reference, a phase value of 0 or 1 means the star is at its brightest, and a value of 0.5 corresponds to when the star is at its darkest.

From our calculated phase values, we can see that HV9 was actually observed right at its maximum light. It can also be seen from our phase calculations that HV7 was observed halfway between minimum and maximum light (leaning towards minimum). Since HV9 was at a brighter phase than HV7, our minimum light theory was incorrect, and there is another reason as to why the quality of HV9's spectra is bad. It’s also important to note that the inconsistencies across the frames for both stars are most likely not because of the time elapsed between exposures, since the first and last frames were taken less than an hour apart, which wouldn't result in a significant change in brightness.

Our mentor, who provided us with the stars' data, also provided us with the galactic location, outer Milky Way halo, and mean magnitudes of the stars. The mean magnitudes of the G filter for HV9 and HV7 are 20.74 and 20.67, respectively. Knowing the magnitude of the stars is very important because fainter stars produce lower-quality spectra. This is caused by a smaller number of photons being collected, resulting in a low S/N ratio and making it difficult to distinguish real stellar features from background noise.

RR Lyrae stars are excellent distance indicators because they are standard candle stars, meaning their absolute magnitude (intrinsic brightness) is known. Because of this, we can calculate the distances of these stars using the relationship between their measured apparent magnitudes and their known absolute magnitudes[4]. The faint magnitudes of HV9 and HV7 suggest that they are located far in the Galactic halo. Therefore, a physical reason for the poor quality of the stars' spectra could simply be that the stars are very far and not very bright.
