To artificially improve the S/N ratio and to help identify the absorption line of H$\alpha$ I applied a smoothing to my frames using IRAF. ## Describe your results
IRAF’s rvidlines velocity calculation was less accurate than the values found through manual calculations. In most of the spectra, the Hα absorption line displayed a small emission-like upturn near the centre of the line. Because rvidlines measures the peak of a line, the centre of the absorption could only be marked as higher or lower than the actual centre. Thus, the velocities were either an overestimate or an underestimate of the velocity. Because of this, I did not use rvidlines velocity output values for my final results. Instead, I used splot on IRAF to determine the centre wavelength of the absorption lines and calculated the velocities manually.

To correct the observed velocities for Earth’s motion, I calculated the heliocentric velocity correction using IRAF’s rvcorrect tool. To be able to use this tool, I first edited the FITS header of one of my reduced science frames to include the observatory name in the correct IRAF recognized format (gemini-south instead of Gemini-south). The heliocentric correction for the observation was found to be 20.81 km/s.

I calculated the final radial velocity for each frame by combining the observed velocity and the heliocentric correction using the following formula

$RVfinal = Vobs + Vcorr + \frac{Vobs \times Vcorr }{c}$


where Vobs is the observed radial velocity, Vcorr is the heliocentric correction, and c is the speed of light in km/s. This formula was provided to my from my mentor, it is the relativistic addition of velocities formula under the assumption that c>>v [4].
